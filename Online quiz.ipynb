```python
#This code is for giving multiple attempts(2) for user to select correct answer
def online_quiz():  
    player_name = input("Enter your name to start the quiz: ")
    score=0
    for question in questions:
        count = 0
        print(question)
        print("Enter your answer (A, B, or C)")
        
        while count < 2:  # Allow up to 2 attempts per question
            answer = input("Give answer: ").strip().capitalize()
            
            user_question=question[0]
            if answer==questions_answers[user_question] or answer==questions_answers[user_question]:
                print("Correct answer!")
                score=score+1
                break 
            else:
                count += 1
                if count < 2:
                    print("Wrong answer! You have one more attempt.")
            if count>=2:
                print("You ran out of your attempts for this question")
        
    
    print(f"\n{player_name}, your final score is {score} out of {len(questions)}.")
    print("Thank you for attempting the quiz. Have a good day!")
    print("If you want to play again type yes")
    user_descision=input("yes/no")
    if user_descision=="yes":
            online_quiz()
    elif user_descision=="no":
            print("bye bye!!")
    else:
            print("not understood your message")
online_quiz()
```
