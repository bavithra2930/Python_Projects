```python
products=[ {"title": "T-Shirt", "description": "Cotton T-shirt with your favorite band logo", "tags": ["clothing", "band", "t-shirt"]}, {"title": "Coffee Mug", "description": "Ceramic mug with a funny quote", "tags": ["kitchen", "drinkware", "quote"]}, {"title": "Notebook", "description": "Lined notebook with a stylish cover", "tags": ["office", "writing", "stationery"]}, ]
```

```python
print(products)
```

```python
a_tag=input("Write the asscociated tag of the product ")   #filtering products in tags only
products_array=[]
print(f"Filter results for {a_tag}")
for i in products:
    input1=i["tags"]    
    for j in input1:
        if j==a_tag:
            products_array.append(i)
if len(products_array)>=1:
    print(products_array)
else:
    print("No products found")
```

```python
a_tag=input("Write the keyword of the product ")   #searching ofr products in title,description and also tags
filtered_products=[]
print(f"Filter results for {a_tag}")
for i in products:
    input1=i["title"]
    input2=i["description"]
    if (a_tag in input1) or (a_tag in input2):
        filtered_products.append(i)
    for j in i["tags"]:
         if (j==a_tag):
                filtered_products.append(i)

if len(products_array)>=1:
    print(filtered_products)
else:
    print("No products found")
```
